```python
"""This program is to read csv file all_uk_retailer.csv and perform Quick Sort to sort all the postcodes and perform binary search on the sorted list."""
# First we import Pandas as pd to read the file using pandas Dataframe.
import pandas as pd
# We read the csv file in the object example.
example = pd.read_csv('all_uk_retailer.csv')

# We append all the postcodes in this list. 
postcodes_list = []

# The for loop below iterrate through each row of the csv file and extracts all the postcode in the list postcodes_list.
for index,row in example.iterrows():
    postcodes_list.append(row.postcode)
#print(postcodes_list)


# Credits : https://www.geeksforgeeks.org/python-program-for-quicksort/
def partition(postcodes_list, low, high):
    """This function helps to find the partition position in this case we take rightmost element.
    _____________________________________________________________________________________________
    INPUT : List of the postcodes, low and high.
    OUTPUT: position from where partition is done."""
    
    # We take rightmost element as our pivot.
    pivot = postcodes_list[high]

    # This is the pointer for greater element.
    i = low - 1

    # With the help of this for loop we traverse through each element and compare it with pivot.
    for j in range(low, high):
        if postcodes_list[j] <= pivot:

            # Now we swap the smaller element with the greater if found using this pointer.
            i = i + 1

            # Swapping element at i with element at j.
            (postcodes_list[i], postcodes_list[j]) = (postcodes_list[j], postcodes_list[i])

    # Swap the pivot element with the greater element specified by i.
    (postcodes_list[i + 1], postcodes_list[high]) = (postcodes_list[high], postcodes_list[i + 1])

    # Return the position from where partition is done.
    return i + 1



def quickSort(postcodes_list, low, high):
    """This is the function to perform quicksort. 
    ______________________________________________
    INPUT : List of postcodes, low and high.
    OUTPUT: Returns sorted list of postcode."""
    
    if low < high:
        # Find the pivot element such that elements smaller than pivot are on left and elements greater are on right.
        pi = partition(postcodes_list, low, high)

        # Recursive call on the left of pivot. 
        quickSort(postcodes_list, low, pi - 1)

        # Recursive call on the right of pivot.
        quickSort(postcodes_list, pi + 1, high)



size = len(postcodes_list)
quickSort(postcodes_list, 0, size - 1)        
#print("Sorted postcodes:")
#print(postcodes_list)


#postcodes_list.clear()
```

```python
def binary_search(postcodes_list, low, high, x):
    """This is the function to perform binary search on a sorted list of postcodes.
    __________________________________________________________________________________
    INPUT : List of postcodes, minimum index, maximum index, and value to be searched.
    OUTPUT: To return value as mid if present."""
    # Compare the base case.
    if high >= low:
        # Set the value of mid.
        mid = (high + low) // 2
 
        # If the target is at mid return mid.
        if postcodes_list[mid] == x:
            return postcodes_list[mid]
 
        # If target is smaller than mid then it can only be on the left side of the array.
        elif postcodes_list[mid] > x:
            return binary_search(postcodes_list, low, mid - 1, x)
 
        # Else the target can only be present in right side of the array.
        else:
            return binary_search(postcodes_list, mid + 1, high, x)
    # Return -1 if not present in the array.
    else:
        return -1

# Target 
target = "nw10"
print("THE SIMILAR POSTCODES ARE:")
# For loop helps to iterrate through each value in the sorted list of postcodes.
for x in postcodes_list:
    
    # If target matches the value of x in the loop. 
    if x[0:4] == target.upper():
        # Call the method binary search with appropriate parameters.
        result = binary_search(postcodes_list, 0, len(postcodes_list)-1, x)
        r = str(result)
        print(r) 
```
